### Cleaning Data

```python
datetime_columns = ["dteday"]
for column in datetime_columns:
    day_df[column] = pd.to_datetime(day_df[column])
    
day_df.info()
```

```python
day_df.season.replace(1, "Springer", inplace=True)
day_df.season.replace(2, "Summer", inplace=True)
day_df.season.replace(3, "Fall", inplace=True)
day_df.season.replace(4, "Winter", inplace=True)

day_df.yr.replace(0, 2011, inplace=True)
day_df.yr.replace(1, 2012, inplace=True)

day_df.holiday.replace(0, "No", inplace=True)
day_df.holiday.replace(1, "Yes", inplace=True)

day_df.weekday.replace(0, "Sunday", inplace=True)
day_df.weekday.replace(1, "Monday", inplace=True)
day_df.weekday.replace(2, "Tuesday", inplace=True)
day_df.weekday.replace(3, "Wenesday", inplace=True)
day_df.weekday.replace(4, "Thursday", inplace=True)
day_df.weekday.replace(5, "Friday", inplace=True)
day_df.weekday.replace(6, "Saturday", inplace=True)

day_df.workingday.replace(0, "No", inplace=True)
day_df.workingday.replace(1, "Yes", inplace=True)

day_df.weathersit.replace(1, "Clear", inplace=True)
day_df.weathersit.replace(2, "Mist", inplace=True)
day_df.weathersit.replace(3, "Light Snow", inplace=True)
day_df.weathersit.replace(4, "Heavy Rain", inplace=True)
day_df.head()
```

```python
day_df["temp_group"] = day_df.temp.apply(lambda x: "Low_Temp" if x <= 0.33 else ("High_Temp" if x > 0.66 else "Normal_Temp"))
day_df.groupby(by="temp_group").instant.nunique().sort_values(ascending=False)

day_df["atemp_group"] = day_df.atemp.apply(lambda x: "Low_Atemp" if x <= 0.33 else ("High_Atemp" if x > 0.66 else "Normal_Atemp"))
day_df.groupby(by="atemp_group").instant.nunique().sort_values(ascending=False)

day_df["hum_group"] = day_df.hum.apply(lambda x: "Low_Hum" if x <= 0.33 else ("High_Hum" if x > 0.66 else "Normal_Hum"))
day_df.groupby(by="hum_group").instant.nunique().sort_values(ascending=False)

day_df["windspeed_group"] = day_df.windspeed.apply(lambda x: "Low_windspeed" if x <= 0.33 else ("High_windspeed" if x > 0.66 else "Normal_windspeed"))
day_df.groupby(by="windspeed_group").instant.nunique().sort_values(ascending=False)

day_df.head()
```

```python
datetime_columns = ["dteday"]
for column in datetime_columns:
    hour_df[column] = pd.to_datetime(hour_df[column])
    
hour_df.info()
```

```python
hour_df.season.replace(1, "Springer", inplace=True)
hour_df.season.replace(2, "Summer", inplace=True)
hour_df.season.replace(3, "Fall", inplace=True)
hour_df.season.replace(4, "Winter", inplace=True)

hour_df.yr.replace(0, 2011, inplace=True)
hour_df.yr.replace(1, 2012, inplace=True)

hour_df.holiday.replace(0, "No", inplace=True)
hour_df.holiday.replace(1, "Yes", inplace=True)

hour_df.weekday.replace(0, "Sunday", inplace=True)
hour_df.weekday.replace(1, "Monday", inplace=True)
hour_df.weekday.replace(2, "Tuesday", inplace=True)
hour_df.weekday.replace(3, "Wenesday", inplace=True)
hour_df.weekday.replace(4, "Thursday", inplace=True)
hour_df.weekday.replace(5, "Friday", inplace=True)
hour_df.weekday.replace(6, "Saturday", inplace=True)

hour_df.workingday.replace(0, "No", inplace=True)
hour_df.workingday.replace(1, "Yes", inplace=True)

hour_df.weathersit.replace(1, "Clear", inplace=True)
hour_df.weathersit.replace(2, "Mist", inplace=True)
hour_df.weathersit.replace(3, "Light Snow", inplace=True)
hour_df.weathersit.replace(4, "Heavy Rain", inplace=True)
hour_df.head()
```

```python
hour_df["temp_group"] = hour_df.temp.apply(lambda x: "Low_Temp" if x <= 0.33 else ("High_Temp" if x > 0.66 else "Normal_Temp"))
hour_df.groupby(by="temp_group").instant.nunique().sort_values(ascending=False)

hour_df["atemp_group"] = hour_df.atemp.apply(lambda x: "Low_Atemp" if x <= 0.33 else ("High_Atemp" if x > 0.66 else "Normal_Atemp"))
hour_df.groupby(by="atemp_group").instant.nunique().sort_values(ascending=False)

hour_df["hum_group"] = hour_df.hum.apply(lambda x: "Low_Hum" if x <= 0.33 else ("High_Hum" if x > 0.66 else "Normal_Hum"))
hour_df.groupby(by="hum_group").instant.nunique().sort_values(ascending=False)

hour_df["windspeed_group"] = hour_df.windspeed.apply(lambda x: "Low_windspeed" if x <= 0.33 else ("High_windspeed" if x > 0.66 else "Normal_windspeed"))
hour_df.groupby(by="windspeed_group").instant.nunique().sort_values(ascending=False)

hour_df["hr_group"] = hour_df.hr.apply(lambda x: "Morning" if x >= 4 and x <= 10 else ("Noon" if x > 10 and x <= 15 else ("Afternoon" if x > 15 and x <= 18 else "Night")))
hour_df.groupby(by="hr_group").instant.nunique().sort_values(ascending=False)

hour_df.head()
```

```python
hour_df["temp_group"] = hour_df.temp.apply(lambda x: "Low_Temp" if x <= 0.33 else ("High_Temp" if x > 0.66 else "Normal_Temp"))
hour_df.groupby(by="temp_group").instant.nunique().sort_values(ascending=False)

hour_df["atemp_group"] = hour_df.atemp.apply(lambda x: "Low_Atemp" if x <= 0.33 else ("High_Atemp" if x > 0.66 else "Normal_Atemp"))
hour_df.groupby(by="atemp_group").instant.nunique().sort_values(ascending=False)

hour_df["hum_group"] = hour_df.hum.apply(lambda x: "Low_Hum" if x <= 0.33 else ("High_Hum" if x > 0.66 else "Normal_Hum"))
hour_df.groupby(by="hum_group").instant.nunique().sort_values(ascending=False)

hour_df["windspeed_group"] = hour_df.windspeed.apply(lambda x: "Low_windspeed" if x <= 0.33 else ("High_windspeed" if x > 0.66 else "Normal_windspeed"))
hour_df.groupby(by="windspeed_group").instant.nunique().sort_values(ascending=False)

hour_df["hr_group"] = hour_df.hr.apply(lambda x: "Morning" if x >= 4 and x <= 10 else ("Noon" if x > 10 and x <= 15 else ("Afternoon" if x > 15 and x <= 18 else "Night")))
hour_df.groupby(by="hr_group").instant.nunique().sort_values(ascending=False)

hour_df.head()
```
